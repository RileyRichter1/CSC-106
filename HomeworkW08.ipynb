### 1. Normal function. (25 points)

```python
import math

z = 9

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

print(sigmoid(z))
```

### 2. Lambda function.(25 points)

```python
import math

sigmoid = lambda x: 1 / (1 + math.exp(-x))

print(sigmoid(z))
```

### 3. Generate linearly spaced values from -10 to 10 to plot the function

```python
import numpy as np
import matplotlib.pyplot as plt

sigmoid = lambda x: 1 / (1 + np.exp(-x))

x_values = np.linspace(start=-10, stop=10, num=100)
sigmoid_values = sigmoid(x_values)

plt.plot(x_values, sigmoid_values)
plt.title('Sigmoid Function Plot')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.show()



```
